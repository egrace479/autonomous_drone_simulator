# Use this script to map flight paths 

```python
# import required libraries
import folium
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
import pandas as pd
```

```python
# load data
flight = pd.read_csv("/users/PAS2136/kline377/kenya_drones/simulator/data/DJI_0063/DJI_0063_flight_record.csv")
```

```python
# create a fence around the flight area
fence = (flight['latitude'].min()-0.001, flight['longitude'].min()-0.001, flight['latitude'].max()+0.001, flight['longitude'].max()+0.001)
avg_lat = (fence[0] + fence[2])/2
avg_lon = (fence[1] + fence[3])/2

# create map
mapit=None
m = folium.Map(location= [avg_lat, avg_lon], zoom_start=18, control_scale=True)
folium.Rectangle(bounds=[(fence[0], fence[1]), (fence[2], fence[3])], color='#3186cc', fill=False).add_to(m)
for index, row in flight.iterrows():
    folium.CircleMarker([row['latitude'], row['longitude']], radius=1, color='red').add_to(m)
folium.Marker([flight['latitude'].iloc[0], flight['longitude'].iloc[0]], popup='Start', icon=folium.Icon(color='green')).add_to(m)
folium.Marker([flight['latitude'].iloc[-1], flight['longitude'].iloc[-1]], popup='End', icon=folium.Icon(color='blue')).add_to(m)
m # output map
```

```python
# save map
m.save('visualizations/flight_map.html')
```
